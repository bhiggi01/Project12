## Train Data

First, load the images with the "cv2" library with 350x350x3 size and normalize the images so they are all similar. 

The Train Data is split into a Train and Validation dataset with a 80:20 ratio.

As the test data is not labeled, the validation data is essential to test the performance of the models.

```python
# Set the image size to 350*350 # We do not need to set the 3 channels as this is done by default.
image_width, image_height = 350, 350

# Train directory
train_dir = "C:/Users/Kolobane/OneDrive/CIT MSc Data Science Modules/COMP9060 Applied Machine Learning/Assignment 3/Data/train/"

# Empty lists
images = [] # images themselves
labels = [] # Labels needed for classfication

# Iterate though each file and set the image size and give classification from the file name
for filename in os.listdir(train_dir):
    if filename.endswith(".jpg"):
        
        # Load each file
        img = cv2.imread(os.path.join(train_dir, filename)) # Notr. cv2 automatically reads in as 3 channels.
        
        # CV2 uses BGR but this will give a blue tinge when I use Matplotlib or seaborn so I will convert to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Resize the image to teh 350*350
        img = cv2.resize(img, (image_width, image_height))
        
        images.append(img)
        
        # Set the label based on the file name
        if "cat" in filename:
            labels.append("cat")
        else:
            labels.append("dog")

# Converts the images to numpy arrays # Necessary for training models as libraries need images in numpy arrasys
images = np.array(images)
labels = np.array(labels)
        
# This normalizes the data and helps to speed up training.
images = images.astype("float32")/ 255.0

# Encode labels
le = LabelEncoder()
labels = le.fit_transform(labels)

# Save the images and labels files
np.save("preprocessed_images.npy", images)
np.save("preprocessed_labels.npy", labels)
```

### Create custom function to test CNN Hyperparameters

```python
# create train genertor
train_data_generator = ImageDataGenerator(rescale=1./255)

# Flow the data in batches
train_generator = train_data_generator.flow(
    train_images, train_labels,
    batch_size=32,
)
```

```python
validation_data_generator = ImageDataGenerator(rescale=1./255)

# Generate batches of preprocessed data from the validation set
validation_generator = validation_data_generator.flow(
    val_images,
    val_labels,
    batch_size=32,
)
```

```python
def create_combined_model(num_conv_layers, num_filters, num_dense_layers, num_nodes):
    model = Sequential()
    
    # Input a first Conv2D layer
    model.add(Conv2D(num_filters, (3,3), activation="relu", input_shape=(350,350,3)))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    filters = num_filters *2
    
    for _ in range(1, num_conv_layers):
        model.add(Conv2D(filters, (3,3), activation="relu"))  
        model.add(MaxPooling2D(pool_size=(2,2)))
        filters *=2
        
    model.add(Flatten())
    model.add(Dropout(0.5))
    
    nodes = num_nodes
    for _ in range(num_dense_layers):
        model.add(Dense(nodes, activation="relu"))
        nodes = nodes // 2
        
    model.add(Dense(1, activation="sigmoid")) # binary classification so output a single probability
    
    # COmplile the model
    model.compile(loss="binary_crossentropy",
                      optimizer="adam",
                      metrics=["accuracy"])
    return model
```

```python
def find_best_hyperparameters():
    
    num_epochs = 3
        
    best_accuracy =0
    best_hyperparameters = None
    
    for num_conv_layers in range(3,6):
        for num_filters in [16, 32]: # Need more filters later
            for num_dense_layers in range (1,3): # Need to add more range
                for num_nodes in [64,128]: # and more nodes
                    
                    model = create_combined_model(num_conv_layers, num_filters, num_dense_layers, num_nodes)
                    
                    # Train the model
                    history = model.fit_generator(
                        train_generator,
                        validation_data= validation_generator,
                        epochs=num_epochs,
                        verbose=1)
                    
                    # Evaluate the model
                    test_loss, test_accuracy = model.evaluate_generator(validation_generator, verbose=0)
                    
                    print(f"Model with {num_conv_layers} conv layers, {num_filters} initial filters, "
                          f"{num_dense_layers} dense layers, {num_nodes} nodes - Test Accuracy: {test_accuracy}")
                    
                    if test_accuracy > best_accuracy:
                        best_accuracy = test_accuracy
                        best_hyperparameters = (num_conv_layers, num_filters, num_dense_layers, num_nodes)
                        
    return best_hyperparameters, best_accuracy

best_hyperparameters, best_accuracy = find_best_hyperparameters()

best_num_conv_layers, best_num_filters, best_num_dense_layers, best_num_nodes = best_hyperparameters


print(f"Best hyperparameters:")
print(f"Number of convolutional layers: {best_num_conv_layers}")
print(f"Initial Number of filters: {best_num_filters}")
print(f"Number of dense layers: {best_num_dense_layers}")
print(f"Number of nodes in dense layers: {best_num_nodes}")
print(f"Best test accuracy: {best_accuracy}")
```
